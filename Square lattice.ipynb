```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

# Defining constants and subfunctions

```python
a_ = 10 # atom separation
a = a_ # horizontal separation between hexagons/unit cells
v1 = np.array([a , 0])
v2 = np.array([0 , a])
def R_sq(m,n):
    r = m*v1 + n*v2
    rv = r
    return rv      # Location vector of the unit cells/atoms in 2D with m,n = 0,1,2,3...
```

# Data production

```python
mmax = 10
nmax = 10
m = np.arange(0,mmax)
n = np.arange(0,nmax)
R_valx = []
R_valy = []
for m1 in m:
    for n1 in n:
        r = R_sq(m1,n1)
        rx = r[0]
        ry = r[1]
        R_valx = np.append(R_valx, rx)
        R_valy = np.append(R_valy, ry)
```

# Plotting

## Sites

```python
plt.figure()
plt.scatter(R_valx, R_valy, s=20, c= 'r', label="Atomic sites")
plt.xlabel("x position (nm)"); plt.ylabel("y position (nm)"); plt.title("Atoms in a Square Lattice")
plt.legend()
plt.grid()
plt.show()
```

## Arrows

```python
def arrows_sq_nn(mmax, nmax, s):
    x_centre = mmax//2
    y_centre = nmax//2
    c0 = [x_centre, y_centre]
    da1 = [s, 0]
    da2 = [0, s]
    da3 = [-s, 0]
    da4 = [0, -s]
    w = 0.01
    plt.arrow(*c0, *da1, color = 'g', ls = '--', width = w, length_includes_head = True, head_width = 10*w, head_length = 4*w)
    plt.arrow(*c0, *da2,color = 'g', ls = '--', width = w, length_includes_head = True, head_width = 10*w, head_length = 4*w)
    plt.arrow(*c0, *da3,color = 'g', ls = '--', width = w, length_includes_head = True, head_width = 10*w, head_length = 4*w)
    plt.arrow(*c0, *da4,color = 'g', ls = '--', width = w, length_includes_head = True, head_width = 10*w, head_length = 4*w)
```

```python
arrows_sq_nn(10,10,1)
```

# The full function

```python
def sq_lat(a_, mmax, nmax, ms):
    a_ = 10 # atom separation
    a = a_ # horizontal separation between hexagons/unit cells
    v1 = np.array([a , 0])
    v2 = np.array([0 , a])
    def R_sq(m,n):
        r = m*v1 + n*v2
        rv = r
        return rv      # Location vector of the unit cells/atoms in 2D with m,n = 0,1,2,3...
    m = np.arange(0,mmax)
    n = np.arange(0,nmax)
    R_valx = []
    R_valy = []
    for m1 in m:
        for n1 in n:
            r = R_sq(m1,n1)
            rx = r[0]
            ry = r[1]
            R_valx = np.append(R_valx, rx)
            R_valy = np.append(R_valy, ry)
    plt.figure()
    plt.scatter(R_valx, R_valy, s=20, c= 'r', label="Atomic sites", zorder = 5)
    for m1 in m:
        for n1 in n[0:-1]:
            j = m1*nmax
            if m1 == 0:
                if n1 ==0:
                    plt.plot((R_valx[n1+j],R_valx[n1+j+1]),(R_valy[n1+j],R_valy[n1+j+1]), c = 'k', label = 'Bonds')
                else:
                    plt.plot((R_valx[n1+j],R_valx[n1+j+1]),(R_valy[n1+j],R_valy[n1+j+1]), c = 'k')
            else:
                plt.plot((R_valx[n1+j],R_valx[n1+j+1]),(R_valy[n1+j],R_valy[n1+j+1]), c = 'k')
    for m1 in m[0:-1]:
        for n1 in n:
            j = m1*nmax
            j_ = (m1+1)*nmax
            plt.plot((R_valx[n1+j],R_valx[n1+j_]),(R_valy[n1+j],R_valy[n1+j_]), c = 'k')
    plt.xlabel("x position (nm)"); plt.ylabel("y position (nm)"); plt.title("Atoms in a Square Lattice")
    plt.legend(loc = 'upper left')
    x_centre = mmax*a_//2
    y_centre = nmax*a_//2
    s = a_
    c0 = [x_centre, y_centre]
    da1 = [s, 0]
    da2 = [0, s]
    da3 = [-s, 0]
    da4 = [0, -s]
    w = 0.05
    plt.grid()
    plt.arrow(*c0, *da1, color = 'y', ls = '--', width = w, length_includes_head = True, head_width = 10*w, head_length = 4*w, zorder = 10)
    plt.arrow(*c0, *da2, color = 'y', ls = '--', width = w, length_includes_head = True, head_width = 10*w, head_length = 4*w, zorder = 10)
    plt.arrow(*c0, *da3, color = 'y', ls = '--', width = w, length_includes_head = True, head_width = 10*w, head_length = 4*w, zorder = 10)
    plt.arrow(*c0, *da4, color = 'y', ls = '--', width = w, length_includes_head = True, head_width = 10*w, head_length = 4*w, zorder = 10)
    plt.show()
```

# Execution
